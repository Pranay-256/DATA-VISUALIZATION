## histplot() â€“ Basic Histogram

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.histplot(tips["total_bill"], bins = 50, kde = True)#,color = "green")
plt.title("Histogram of total bills")
```

## kdeplot() â€“ Kernel Density Estimate

```python
sns.kdeplot(tips["total_bill"], fill = True, color = "green")
```

## rugplot() â€“ Small Ticks for Observations

```python
sns.rugplot(x = tips["total_bill"],height = 0.2, color = "brown")
```

## Multiple Distributions with hue

```python
sns.histplot(x = tips["total_bill"], hue = tips["sex"], kde = True, bins = 20,
palette = "Set2")
```
