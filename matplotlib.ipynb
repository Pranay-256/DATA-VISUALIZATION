###Plotting Two Arrays



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

a = [1,2,3,4,5]
b = [10, 20, 30, 40, 50]

plt.plot(a, b)

plt.xlabel("counter")
plt.ylabel("marks")
plt.title("diagram")
```

###Figure

```python
x = [1,2,3,4,5]
y = [10, 20, 30, 40, 50]
plt.figure(figsize = (6,4))
plt.plot(x,y)
plt.plot(y,x)
plt.xlabel("age")
plt.ylabel("score")
plt.grid(True)
plt.title("Random Diagram")
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]
plt.plot(x, y1)
plt.plot(x, y2)
```

###subplots

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize =(12, 10))
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.xlabel("Age Comparison")

plt.subplot(2, 2, 4)
#plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.xlabel("Salary Comparison")
```

###object oriented interface

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

fig = plt.figure(figsize = (8, 6))#object = class

ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax.plot(x, y1)


ax2.plot(x, y2)
```

###figure size, aspect ratio and dpi

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize = (8, 6), dpi = 120)
plt.plot(x, y)
```

###Legends, label and title

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.figure(figsize = (7, 4))

# plot first line
plt.plot(x, y1, color = "blue", marker = "o", label = "square")

# plot second line
plt.plot(x, y2, color = "red", marker = "s", label = "Line")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Example: Legends, Labels & Title")
plt.legend(loc = "upper left")
```

###plot range(axis limits)

```python
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

fig, ax = plt.subplots(figsize = (7, 4))

ax.plot(x, y, color = "teal", marker = "o", linewidth = 2)

ax.set_xlim(-1, 6)
ax.set_ylim(-5, 30)
```

###scatter, histogram and boxplot

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]
plt.figure(figsize = (12, 4))
plt.subplot(1,3,1)
plt.scatter(x, y)

plt.subplot(1,3,2)
plt.hist(x)

plt.subplot(1,3,3)
plt.boxplot(x)
```
