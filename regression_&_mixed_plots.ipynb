# regplot() â€“ Scatter + Regression Line

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.regplot(x = tips["total_bill"], y = tips["tip"],color = "black")#doesn't supports hue
plt.title("regression plot for tip and total_bill")
```

# lmplot() â€“ Regression with Facets

```python
sns.lmplot(x = "total_bill", y = "tip", hue = "sex", data = tips, height = 5, aspect = 1.2)
```

# jointplot() â€“ Scatter + Histogram / KDE

```python
#sns.jointplot(x = tips["total_bill"], y = tips["tip"],kind = "scatter")
sns.jointplot(x = tips["total_bill"], y = tips["tip"],kind = "kde", fill = True)
```
