## You can load sample DataSets using SNS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


iris = sns.load_dataset("iris")
iris
tips = sns.load_dataset("tips")
penguines = sns.load_dataset("penguins")
tips
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"])
plt.title("Scatter PLot showing total bill vs total tips")
```

```python
tips["total_bill"].corr(tips["tip"])
```

```python
sns.lineplot(x = tips["day"], y = tips["tip"], estimator = "mean", marker = "o")
```

```python
sns.set_style("darkgrid") #options: white, dark, darkgrid, whitegrid, ticks
sns.set_palette("bright") #others: deep, bright, dark, muted, colorblind

sns.scatterplot(x = tips["total_bill"], y = tips["tip"], hue = tips["sex"])
```
